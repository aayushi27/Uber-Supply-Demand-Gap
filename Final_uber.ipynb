```python
# creating a time slot for various time period of day

def time_period(x):
    'divide the day time into four categories'
    if x < 5:
        return "Pre_Morning"
    elif 5 <= x < 10:
        return "Morning_Rush"
    elif 10 <= x < 17:
        return "Day_Time"
    elif 17 <= x < 22:
        return "Evening_Rush"
    else:
        return "Late_Night"

uber_df['time_slot'] = uber_df.req_hour.apply(lambda x: time_period(x))
```

```python
# We know that there is gap between demand and supply of cabs at both the places but in different time slots

# Hence we will find the gap between the demand and supply

uber_df['Total Booking'] = 'Total Booking'

uber_df_demand = uber_df.groupby(['Booking Status', 'time_slot'])['Request id'].count()

uber_df_supply = uber_df.groupby(['Total Booking', 'time_slot'])['Request id'].count()
uber_df_supply.columns = ['Booking Status', 'time_slot', 'Request_id']
uber_df_demand_supply = uber_df_demand.append(uber_df_supply)
uber_df_demand_supply = uber_df_demand_supply.reset_index()
print(uber_df_demand_supply)
```

```python
# Giving the pictorial representation for the above calculation

plt.figure(figsize=(10,5))
sns.set_context('talk')
sns.barplot(x='time_slot', y ='Request id', hue = 'Booking Status', data = uber_df_demand_supply)
plt.ylabel('Booking Requests')
plt.legend(fontsize=10, frameon=True)
plt.show()
```

```python
# create pie charts instead of a count plots
def pie_chart(dataframe):
    """
    creates a pie chart
    input: dataframe with a 'category' as index and a numerical column
    output: pie chart
    """
    labels = dataframe.index.values
    sizes = dataframe['Status'].values
        
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```

```python
# Status of trips @ Evening Rush where pickup point is Airport
uber_df_airport = uber_df.loc[(uber_df["Pickup point"] == "Airport") & (uber_df.time_slot == "Evening_Rush")] 
uber_df_airport_count = pd.DataFrame(uber_df_airport.Status.value_counts())
pie_chart(uber_df_airport_count)
```

```python
supply_df = uber_df[uber_df['Status'] == 'Trip Completed'].groupby(['time_slot','Pickup point'])['Request id'].count().reset_index().rename(columns = {'Request id':'Supply'}).copy()
```

```python
demand_df = uber_df.groupby(['time_slot','Pickup point'])['Request id'].count().reset_index().rename(columns = {'Request id':'Demand'}).copy()
```

```python
demand_supply_df = demand_df.merge(supply_df,on=['time_slot','Pickup point'])
```

```python
demand_supply_df['Gap'] = demand_supply_df['Demand']-demand_supply_df['Supply']
```

```python
demand_supply_df.head()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x = 'time_slot',y= 'Gap',data = demand_supply_df)
plt.show()
```

```python
print('Average gap across different timeslots for all the days:','\n')
print(demand_supply_df.groupby('time_slot')['Gap'].sum().transform(lambda x:x/5))
```

```python
print('Average gap across different timeslots for all the days:','\n')
print(demand_supply_df.groupby(['time_slot','Pickup point'])['Gap'].mean())
```

```python
# Hence time_slot when the highest gap exist is for 'Evening_Rush'

plt.figure(figsize=(10,5))
sns.barplot(x='time_slot', y ='Gap', hue = 'Pickup point', data = demand_supply_df)
plt.show()
```
